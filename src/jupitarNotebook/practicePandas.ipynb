```python
all_symptoms = set()
for col in df.columns:
    unique_vals = set(df[col].unique())
    unique_vals.discard(np.nan)      # removes NaN if present
    unique_vals.discard('')          # removes empty string
    all_symptoms = all_symptoms.union(unique_vals)
```
