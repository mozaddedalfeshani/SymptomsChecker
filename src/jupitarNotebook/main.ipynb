```python
# Filter rows for a disease
panic_df = df[df['diseases'] == 'panic disorder']

# Drop the 'diseases' column and sum the symptom occurrences
panic_symptom_counts = panic_df.drop('diseases', axis=1).sum().sort_values(ascending=False).head(10)

# Plot top symptoms
panic_symptom_counts.plot(kind='barh', color='orange')
plt.title("Top Symptoms in Panic Disorder")
plt.xlabel("Frequency")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Total sum of each symptom
total_symptom_counts = df.drop('diseases', axis=1).sum().sort_values(ascending=False).head(15)

# Bar plot
total_symptom_counts.plot(kind='bar', color='green',  figsize=(14,7))
plt.title("Most Common Symptoms Across All Diseases")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Install seaborn if not already installed
%pip install seaborn

import seaborn as sns

# Take a sample of symptom columns
sample_symptoms = df.drop('diseases', axis=1).iloc[:, :20]  # First 20 symptoms

# Correlation matrix
correlation = sample_symptoms.corr()

# Plot heatmap
plt.figure(figsize=(17, 7))
sns.heatmap(correlation, cmap='coolwarm', annot=False)
plt.title("Symptom Correlation Heatmap")
plt.tight_layout()
plt.show()

```
